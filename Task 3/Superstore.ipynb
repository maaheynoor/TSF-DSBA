### _Correlation between each numerical attribute of dataset_

```python
df.corr()
```

```python
fig,axes = plt.subplots(1,1,figsize=(9,6))
sb.heatmap(df.corr(), annot= True)
plt.show()
```

Sales and Profit show a correlation of 0.48

### _Pair Plot_

```python
sb.pairplot(df)
```

Correlation exists only between Profit and Sales attributes of dataset

### _Scatter Plot Sales vs Profit with Discount_

```python
df.plot(kind = "scatter", figsize = (16,6), x = "Sales", y ="Profit" ,c ="Discount", s =20, marker ="o",colormap ="viridis")
plt.ylabel("Sales",fontsize =13)
plt.ylabel("Profits",fontsize =13)
plt.title("Interdependence of Sales, Profit, and Discount")
plt.show()
```

For lower values of discount there is a positive correlation between Profit and Sales

For higher discount rates, the corretion becomes negative. i.e. Profit decreases as Sales increase

### _Visualize Categories and Sub-Categories_

```python
plt.figure(figsize=(4,8))
plt.scatter('Category','Sub-Category', data=df)
plt.show()
```

### _Visualize counts of categories and sub categories using Pie plot_

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,8))
ax1.pie(df['Category'].value_counts(),autopct="%1.1f%%",labels=df['Category'].unique(),
        colors=['#99e699','#0088cc','#ff6699'])

ax2.pie(df['Sub-Category'].value_counts(),autopct="%1.1f%%",labels=df['Sub-Category'].unique())

plt.show()
```

### _Bar Graph for Sub-Category_

```python
plt.figure(figsize=(16,4))
sb.countplot(x=df['Sub-Category'])
plt.show()
```

Binders are bought most frequently

### _Total Sales for each Sub-Category_

```python
plt.figure(figsize=(16,6))
df.groupby('Sub-Category')['Sales'].sum().plot.bar()
plt.show()
```

Sales of Chairs and Phones is quite large compared to other sub categories

### _Total Profit for each Sub-Category_

```python
plt.figure(figsize=(16,6))
df.groupby('Sub-Category')['Profit'].sum().plot.bar()
ax = plt.gca()
ax.axhline(0, color="gray")
plt.show()
```

Maximum profit is observed for Copiers while Tables have the maximum loss

### _Total Discount for each Sub-Category_

```python
plt.figure(figsize=(16,6))
df.groupby('Sub-Category')['Discount'].sum().plot.bar()
plt.show()
```

Discount for Binders is too high. There is least discount for Copiers (and Maximum Profit as seen above).

### _Bar graph for Sub Category separated region wise_

```python
plt.figure(figsize=(15,8))
sb.countplot(x="Sub-Category", hue="Region", data=df,palette="Set3")
plt.show()
```

Most orders come from the western part of USA and the least number of order is observed in the southern part.

### _Top 10 States based of Sales_

```python
plt.figure(figsize=(16,6))
df.groupby('State')['Sales'].sum().nlargest(n=10).plot.bar()
plt.show()
```

California has the maximum sales

### _Total Profit from each State_

```python
plt.figure(figsize=(16,6))
df.groupby('State')['Profit'].sum().nlargest(n=df['State'].nunique()).plot.bar()
plt.show()
```

California and New Yorks show higher gain in profit. On the other hand, Texas shows a great loss.

## Conclusion
Following weak areas can be considered to improve and make more profit:
* Sales, Profit and Discount are related to each other. To increase the profit, discount should be reduced but this may affect the Sales.
* Sales of Labels and Fasteners are quite low, steps should be taken to improve the sales
* Tables give maximum loss 
* Discount on Binders is very high compared to other sub-categories
* Less number of orders are seen from the South region
* Superstore is running into a loss in the State of Texas 
    
Above mentioned weak areas should be considered and proper steps should be taken to increase the profit.
